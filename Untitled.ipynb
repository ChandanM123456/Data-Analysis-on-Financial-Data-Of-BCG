```python
import pandas as pd

# Load your data
df = pd.read_csv(r"C:\Users\DELL\Desktop\bcg project\financial_data.csv")

# Sort data (optional)
df = df.sort_values(by=['Company', 'Year'])

# Calculate YoY growth
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
df['Asset Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liability Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100

# View summary
print(df)


import matplotlib.pyplot as plt

for company in df['Company'].unique():
    subset = df[df['Company'] == company]
    plt.plot(subset['Year'], subset['Total Revenue'], label=company)

plt.title("Total Revenue Comparison")
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```
